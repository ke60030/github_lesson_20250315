## 這裡是一個建立`正弦波`圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel("角度")
plt.ylabel("sin 值")
plt.title("正弦波圖表")
plt.grid(True)
plt.show()
```

- 餘弦波圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸數據，從 0 到 2π，共 100 個點
x = np.linspace(0, 2*np.pi, 100)

# 計算 y 軸數據，即 cos(x) 的值
y = np.cos(x)

# 使用 plot 函數繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title("餘弦波圖表")
plt.xlabel("角度")
plt.ylabel("cos 值")

# 顯示網格
plt.grid(True)

# 顯示圖表
plt.show()
```

- 同時顯示sin,cos的正弦波和餘弦波

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 軸數據
x = np.linspace(0, 2 * np.pi, 500)

# 計算正弦波和餘弦波
y_sin = np.sin(x)
y_cos = np.cos(x)

# 繪製圖形
plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.plot(x, y_cos, label='cos(x)', color='red')

# 添加標籤和標題
plt.title('sin cos')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

# 顯示圖例
plt.legend()

# 顯示圖形
plt.grid()
plt.show()
```
