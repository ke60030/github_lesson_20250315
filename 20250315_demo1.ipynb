## 這裡是一個建立`正弦波`圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel("角度")
plt.ylabel("sin 值")
plt.title("正弦波圖表")
plt.grid(True)
plt.show()
```
